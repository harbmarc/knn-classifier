# Classification

```python
db_name = 'db1'  # db1 | db2 | cifar-100
distance = "L2"  # L1 | L2
knn_k = 7

assert(knn_k % 2 == 1)

classif_score = 0.

# Load DB
# Return classes as N (class_name), train_data as Nx2 (class_idx, image), test_data as Nx2 (class_idx, image)
classes, train_data, test_data = input_tools.load_db(db_name)
```

```python
# Train classifiers
classifiers = np.zeros_like(classes, dtype=np.object)
for i, c in enumerate(classes):
    print("%s" % c)
    class_data = train_data[train_data[:,0]==i][:,1]
    classifiers[i] = class_data 
```

```python
def knn_test(classifiers, q):
  
    k = knn_k
    distances = np.zeros((len(classes),classifiers[0].shape[0]))
    cl_pts = np.zeros((len(classes),k))
    p = np.zeros(len(classes))

    if distance == "L1":
        distances = [np.sum(np.abs(classifiers[i]-q)) for i in range(len(classes))]
        pass
    elif distance == "L2":
        for i in range(len(classes)):
          distances[i] = [np.sqrt(np.sum((classifiers[i][j]-q)**2)) for j in range(classifiers[i].shape[0])]  
          distances[i].sort()
          cl_pts[i] = distances[i][:k]   

          pass
    else:
        raise NotImplementedError()
    clt_pts = cl_pts.tolist()
    for i in range(k):
      lst2 = [item[0] for item in clt_pts]
      z1 = np.argmin(lst2)
      p[z1] = p[z1]+1 
      clt_pts[z1].pop(0)
    index=np.argmax(p)


    
    return index  # Change this line (of course)

test_data = test_data[:100]  # Select the 100 first
predictions = np.zeros(test_data.shape[0], dtype=np.int)
classif_score= 0 
for i in range(len(test_data)):
    if i % 10 == 0:
        print('!!!! !! Test %d / %d' % (i, len(test_data)))

    label, q = test_data[i]

    predictions[i] = knn_test(classifiers, q)
    if label == predictions[i]:
      classif_score += 1
```

```python
print(classif_score)
classif_score /=len(test_data)
print("Classification score {}" .format(classif_score*100))
output = [np.vstack([test_data[i, 1], classifiers[label_p][0] * ([[[0, 1, 0]]] if label_p == test_data[i, 0] else [[[0, 0, 1]]])]) for i, label_p in enumerate(predictions)]
output_gallery = image_tools.gallery_db(output, 10)
plt.figure('output_gallery')
plt.imshow(output_gallery[..., ::-1])
```
